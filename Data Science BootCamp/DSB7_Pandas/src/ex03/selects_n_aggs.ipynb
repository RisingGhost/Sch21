## 1. Reading Json and setting index

```python
df =pd.read_json("../ex02/auto.json")
df.set_index('CarNumber', inplace = True)
df
```

## 2. Some selections

```python
df.info()
```

- *Fines* > 2100

```python
df.loc[df["Fines"] > 2100]
```

- *Fines* > 2100 and *Refund* = 2

```python
df.loc[(df["Fines"] > 2100) & (df["Refund"] == 2)]
```

- Models ffrom list

```python
df.loc[df["Model"].isin(['Focus', 'Corolla'])]
```

- Numbers from list

```python
df.loc[df.index.isin(['Y7689C197RUS', '92928M178RU ', '7788KT197RU ', 'H115YO163RU ', 'X758HY197RU '])]
```

# 3. Aggregations with make_n_models

```python
df.groupby('Make')['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].median()
```

```python
df.groupby(['Make', 'Model'])['Fines'].count()
```

```python
df.groupby(['Make', 'Model'])['Fines'].agg(['min', 'max'])
```

```python
df.groupby(['Make', 'Model'])['Fines'].std()
```
