## 2. Count obs

```python
df.count()
```

## 3. Drop duplicates

```python
df = df.drop_duplicates(subset=["CarNumber", "Make_n_model", "Fines"], keep='last')
df.count()
```

## 4. Working with missing

- считаем наны

```python
df.isna().sum()
```

- дропаем всё где больше 500 нанов

```python
df = df.dropna(axis = 1, thresh = len(df)-500)
df.isna().sum()
```

- заполняем рефунд через форвард-филл

```python
df["Refund"] = df["Refund"].ffill()
df.isna().sum()
```

- заполняем *Fines* через среднее

```python
df["Fines"] = df["Fines"].fillna(df["Fines"].mean())
df.isna().sum()
df["Fines"]
```

## 5. Splitting Make_n_Model column

```python
df["Make"] = df['Make_n_model'].apply(lambda x: str(x).split(" ")[0])
df["Model"] = df['Make_n_model'].apply(lambda x: str(x).split(" ")[1] if len(str(x).split(" ")) == 2 else "") 
df = df.drop(columns = 'Make_n_model')

```
