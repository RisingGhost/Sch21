```python
import pandas as pd
```

## Reading .log
используемые аргументы для read_csv:
- "delimiter" таб как разделитель (при считывании).
- "skiprows" пропускаем две строки по индексам
- "skipfooter" фильтруем последние 2 строки
- "header" = None чтобы не читать первую строку как заголовок
- "names" названия для столбцов
- "index_cols" для выбора столбца на индексы

```python
df = pd.read_csv("../data/feed-views.log", delimiter="\t", skiprows=[2,3], skipfooter=2, header=None, names = ["datetime", "user"], index_col='datetime')
df
```

## Rename index and output

```python
df.index.name = "date_time"
df.head()
```
