## 2. Сразу к делу (рисуем график)

```python
def draw_box(ax, part, title):
    ctrl = part.loc[part.group=='control', 'diff']
    test = part.loc[part.group=='test',    'diff']
    ax.boxplot(
        [ctrl, test],
        tick_labels=['control','test'],
        showmeans=True,
        boxprops=dict(color='green', linewidth = 2),
        whiskerprops=dict(color='red'),
        capprops=dict(color='red'),
        medianprops=dict(color='red'),
    )
    ax.set_title(title)
    ax.grid(True)
    ax.set_xlabel("[group]")

fig, axes = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
draw_box(axes[0], df[df.time=='after'],  'after')
draw_box(axes[1], df[df.time=='before'], 'before')
fig.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
```
