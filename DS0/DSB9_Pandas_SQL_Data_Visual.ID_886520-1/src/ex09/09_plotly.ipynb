## 1. import an read

```python
import sqlite3
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
db = sqlite3.connect("../data/checking-logs.sqlite")
```

## 2. Preprocessing

```python
data = pd.read_sql("""
            SELECT uid, timestamp, numTrials
            FROM checker WHERE
                status = 'ready'
                AND uid LIKE 'user%'
                AND labname = 'project1'
            """,
            db, parse_dates='timestamp')
data.head()
```

```python
data['timestamp'] = data.timestamp.dt.date
data = data.sort_values(by='timestamp')
```

```python
g_data = data.groupby(['uid', 'timestamp'], as_index = False).numTrials.max()
g_data.head()
dates = g_data['timestamp'].unique()
```

```python
pivot = pd.pivot(data = g_data, index = 'uid', columns = 'timestamp', values = 'numTrials')
pivot.head()
```

```python
last_value = 0
for j in range (len(pivot)):
    for i in range (len(pivot.iloc[j])):
        value = pivot.iloc[j][i]
        if 0 < value and value < 999:
            last_value = int(value)
        else:
            pivot.iloc[j][i] = last_value
    last_value = 0
pivot.head()
```

```python
data = pivot
data = data.reset_index()
data.head()
```

## 3. Creating graph like in reference

```python
num_rows = data.shape[0]
num_cols = data.shape[1]
num_frames = num_cols - 1
xaxis_range = [0,num_frames + 2]
x_init = np.array([1])
```

```python
initial_data = []
for id in data['uid'].unique():
    y_axis = np.array(id)
    initial_data.append(go.Scatter(x =x_init, y = y_axis,mode = "lines",name = id))
initial_max = 600
```

```python
frames = []
for f in range (1,num_frames):
    x_axis = np.arange(1,f+1)
    date = str(data.columns[f])
    curr_data = [] 
    title_names = [] 
    start = "For " + date
    for id in data['uid'].unique():
        y_axis = np.array(data[data.uid == id].values.tolist()[0][1:f+1])
        y_axis[0] = 0
        curr_data.append(go.Scatter(x = x_axis, y = y_axis,mode = "lines+markers", name = id))
        title_names.append(id + ": " + str(y_axis[f-1]) +"commit")
    curr_frame = go.Frame(data = curr_data)
    frames.append(curr_frame)
```

```python
figure = go.Figure(
    data = initial_data,
    layout = {
        "title":"Dynamic of commits per user in project1",
        "xaxis":{"range":xaxis_range, "visible":True, "showline":False},
        "yaxis":{"range": [0, 170], "visible":True, "showline":False},
        "updatemenus":[{"type":"buttons","buttons":[{"method":"animate","label":"play", "args":[None]}]}]
        },
    frames = frames
    )
figure.layout.template='seaborn'
```

```python
figure.show()
```
